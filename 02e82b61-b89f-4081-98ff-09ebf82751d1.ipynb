#### –®–∞–≥ 5. –ü–æ–¥—Å—á—ë—Ç –≤—ã—Ä—É—á–∫–∏

__–ö–æ–º–º–µ–Ω—Ç–∞—Ä–∏–π –æ—Ç —Ä–µ–≤—å—é–µ—Ä–∞ ‚Ññ2__
    
–ù–µ—Ç –ø—Ä–æ–≤–µ—Ä–∫–∏ –Ω–∞ –ø—Ä–æ–ø—É—Å–∫–∏ –≤ —Å–≤–æ–¥–Ω–æ–π —Ç–∞–±–ª–∏—Ü–µ –ø–µ—Ä–µ–¥ —Ä–∞—Å—á–µ—Ç–æ–º –≤—ã—Ä—É—á–∫–∏

```python
print("–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ –ø—Ä–æ–ø—É—Å–∫–æ–≤ –ø–µ—Ä–µ–¥ –∞–≥—Ä–µ–≥–∞—Ü–∏–µ–π:")
print(merged_df[['user_id', 'name', 'age', 'city', 'subscription_type', 'month', 'distance', 'duration']].isna().sum())
```

```python
# 5.1 –°–æ–∑–¥–∞–Ω–∏–µ –¥–∞—Ç–∞—Ñ—Ä–µ–π–º–∞ —Å –∞–≥—Ä–µ–≥–∏—Ä–æ–≤–∞–Ω–Ω—ã–º–∏ –¥–∞–Ω–Ω—ã–º–∏
# –°–Ω–∞—á–∞–ª–∞ –æ–∫—Ä—É–≥–ª—è–µ–º –ø—Ä–æ–¥–æ–ª–∂–∏—Ç–µ–ª—å–Ω–æ—Å—Ç—å –ø–æ–µ–∑–¥–∫–∏ –¥–æ —Å–ª–µ–¥—É—é—â–µ–≥–æ —Ü–µ–ª–æ–≥–æ —á–∏—Å–ª–∞
rides_df['duration'] = np.ceil(rides_df['duration'])

# –°–æ–∑–¥–∞–µ–º –Ω–æ–≤—ã–π –¥–∞—Ç–∞—Ñ—Ä–µ–π–º —Å –ø–æ–º–µ—Å—è—á–Ω–æ–π –∞–≥—Ä–µ–≥–∞—Ü–∏–µ–π
aggregated_data = (
    merged_df.groupby(['user_id', 'name', 'age', 'city', 'subscription_type', 'month'])
    .agg(
        total_distance=('distance', 'sum'),
        total_rides=('distance', 'size'),  # –°—á–∏—Ç–∞–µ–º –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –ø–æ–µ–∑–¥–æ–∫
        total_duration=('duration', 'sum')
    )
    .reset_index()
)
```

__–ö–æ–º–º–µ–Ω—Ç–∞—Ä–∏–π –æ—Ç —Ä–µ–≤—å—é–µ—Ä–∞ ‚Ññ2__

–û—Ç–ª–∏—á–Ω–æ, —á—Ç–æ –∏—Å–ø–æ–ª—å–∑—É–µ—à—å –∏–º–µ–Ω–Ω–æ np.ceil(). –ì—Ä–∞–º–æ—Ç–Ω—ã–π —à–∞–≥

```python
# 5.2 –î–æ–±–∞–≤–ª–µ–Ω–∏–µ —Å—Ç–æ–ª–±—Ü–∞ —Å –ø–æ–º–µ—Å—è—á–Ω–æ–π –≤—ã—Ä—É—á–∫–æ–π

# –û–∫—Ä—É–≥–ª—è–µ–º total_duration –ø–æ—Å–ª–µ –∞–≥—Ä–µ–≥–∞—Ü–∏–∏
aggregated_data['total_duration'] = np.ceil(aggregated_data['total_duration'])

# –§—É–Ω–∫—Ü–∏—è –¥–ª—è —Ä–∞—Å—á–µ—Ç–∞ –≤—ã—Ä—É—á–∫–∏
def calculate_revenue(row):
    start_ride_price = 0
    if row['subscription_type'] == 'ultra':
        start_ride_price = 0  
        subscription_fee = 199  
        minute_price = 6  
    else:
        start_ride_price = 50  
        subscription_fee = 0  
        minute_price = 8  

    revenue = (start_ride_price * row['total_rides'] +
               minute_price * row['total_duration'] +
               subscription_fee)
    
    return revenue

# –ü—Ä–∏–º–µ–Ω—è–µ–º —Ñ—É–Ω–∫—Ü–∏—é –¥–ª—è —Ä–∞—Å—á–µ—Ç–∞ –≤—ã—Ä—É—á–∫–∏
aggregated_data['revenue'] = aggregated_data.apply(calculate_revenue, axis=1)

# –ü—Ä–æ–≤–µ—Ä—è–µ–º —Ä–µ–∑—É–ª—å—Ç–∞—Ç—ã
display(aggregated_data.head())
```

**–ü–æ–¥—Ö–æ–¥:** –î–ª—è —ç—Ç–æ–π —Å–∏—Ç—É–∞—Ü–∏–∏ –º—ã –º–æ–∂–µ–º –∏—Å–ø–æ–ª—å–∑–æ–≤–∞—Ç—å —Ç–µ—Å—Ç –Ω–∞ —Ä–∞–∑–Ω–æ—Å—Ç—å —Å—Ä–µ–¥–Ω–∏—Ö (t-—Ç–µ—Å—Ç –¥–ª—è –∑–∞–≤–∏—Å–∏–º—ã—Ö –≤—ã–±–æ—Ä–æ–∫), –ø–æ—Å–∫–æ–ª—å–∫—É –º—ã —Å—Ä–∞–≤–Ω–∏–≤–∞–µ–º –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –æ–±—Ä–∞—â–µ–Ω–∏–π –¥–æ –∏ –ø–æ—Å–ª–µ –æ–±–Ω–æ–≤–ª–µ–Ω–∏—è –¥–ª—è –æ–¥–Ω–∏—Ö –∏ —Ç–µ—Ö –∂–µ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π.

```python
# –ó–∞–¥–∞–Ω–Ω—ã–µ –ø–∞—Ä–∞–º–µ—Ç—Ä—ã
p = 0.1  # –í–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å –ø—Ä–æ–¥–ª–µ–Ω–∏—è –ø–æ–¥–ø–∏—Å–∫–∏
target = 100  # –¶–µ–ª–µ–≤–æ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –ø—Ä–æ–¥–ª–µ–Ω–∏–π
alpha = 0.05  # –£—Ä–æ–≤–µ–Ω—å –∑–Ω–∞—á–∏–º–æ—Å—Ç–∏

# –ò—â–µ–º –º–∏–Ω–∏–º–∞–ª—å–Ω–æ–µ n
n = 1
while True:
    # –í—ã—á–∏—Å–ª—è–µ–º –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å P(X < 100) –¥–ª—è –±–∏–Ω–æ–º–∏–∞–ª—å–Ω–æ–≥–æ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è
    prob = binom.cdf(target - 1, n, p)
    if prob < alpha:
        break
    n += 1

# –ü–æ—Å—Ç—Ä–æ–µ–Ω–∏–µ –≥—Ä–∞—Ñ–∏–∫–∞ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è
x = np.arange(0, n+1)
pmf = binom.pmf(x, n, p)

plt.figure(figsize=(12, 6))
plt.bar(x, pmf, color='skyblue')
plt.title(f'–ë–∏–Ω–æ–º–∏–∞–ª—å–Ω–æ–µ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ (n={n}, p={p})')
plt.xlabel('–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ –ø—Ä–æ–¥–ª–µ–Ω–∏–π –ø–æ–¥–ø–∏—Å–∫–∏')
plt.ylabel('–í–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å')
plt.axvline(x=target, color='red', linestyle='--', label='–¶–µ–ª—å: 100 –ø—Ä–æ–¥–ª–µ–Ω–∏–π')
plt.legend()
plt.grid()
plt.show()

n  # –ú–∏–Ω–∏–º–∞–ª—å–Ω–æ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –ø—Ä–æ–º–æ–∫–æ–¥–æ–≤
```

```python
# –ó–∞–¥–∞–Ω–Ω—ã–µ –ø–∞—Ä–∞–º–µ—Ç—Ä—ã
n = 1_000_000  # –ö–æ–ª–∏—á–µ—Å—Ç–≤–æ —É–≤–µ–¥–æ–º–ª–µ–Ω–∏–π
p = 0.4  # –í–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å –æ—Ç–∫—Ä—ã—Ç–∏—è —É–≤–µ–¥–æ–º–ª–µ–Ω–∏—è
mu = n * p  # –°—Ä–µ–¥–Ω–µ–µ
sigma = np.sqrt(n * p * (1 - p))  # –°—Ç–∞–Ω–¥–∞—Ä—Ç–Ω–æ–µ –æ—Ç–∫–ª–æ–Ω–µ–Ω–∏–µ

# –û—Ü–µ–Ω–∫–∞ –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–∏ P(X <= 399500)
x = 399_500
z = (x - mu) / sigma  # –°—Ç–∞–Ω–¥–∞—Ä—Ç–Ω–æ–µ –Ω–æ—Ä–º–∞–ª—å–Ω–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ
prob = norm.cdf(z)  # –í–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å

# –ü–æ—Å—Ç—Ä–æ–µ–Ω–∏–µ –≥—Ä–∞—Ñ–∏–∫–∞ –Ω–æ—Ä–º–∞–ª—å–Ω–æ–≥–æ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è
x_values = np.linspace(mu - 4*sigma, mu + 4*sigma, 1000)
y_values = norm.pdf(x_values, mu, sigma)

plt.figure(figsize=(12, 6))
plt.plot(x_values, y_values, color='skyblue', label='–ù–æ—Ä–º–∞–ª—å–Ω–æ–µ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ')
plt.fill_between(x_values, y_values, where=(x_values <= x), color='lightcoral', alpha=0.5, label='P(X ‚â§ 399500)')
plt.axvline(x=x, color='red', linestyle='--', label='399500 –æ—Ç–∫—Ä—ã—Ç–∏–π')
plt.title('–ù–æ—Ä–º–∞–ª—å–Ω–∞—è –∞–ø–ø—Ä–æ–∫—Å–∏–º–∞—Ü–∏—è –±–∏–Ω–æ–º–∏–∞–ª—å–Ω–æ–≥–æ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è')
plt.xlabel('–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ –æ—Ç–∫—Ä—ã—Ç–∏–π —É–≤–µ–¥–æ–º–ª–µ–Ω–∏–π')
plt.ylabel('–ü–ª–æ—Ç–Ω–æ—Å—Ç—å –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–∏')
plt.legend()
plt.grid()
plt.show()

prob  # –í–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å
```
